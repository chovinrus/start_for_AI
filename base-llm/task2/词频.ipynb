```python
import jieba
import os

def get_dict_freq(word):
    # 获取 jieba 词典路径
    dict_path = os.path.join(os.path.dirname(jieba.__file__), 'dict.txt')
    
    with open(dict_path, 'r', encoding='utf-8') as f:
        for line in f:
            parts = line.strip().split()
            if len(parts) >= 2 and parts[0] == word:
                return int(parts[1])
    return 0  # 词典中未找到

# 使用示例
print(f"'南京' 的词频: {get_dict_freq('南京')}")
print(f"'北京' 的词频: {get_dict_freq('北京')}")
```

```python
jieba.add_word('比比拉布',freq=10000) #add_word只是改变当前加载的分词器对象而非持久化的词典文件
jieba.add_word('我的刀盾',freq=10000)
jieba.add_word('歪比巴卜',freq=1)
jieba.add_word('我的',freq=20000)
jieba.add_word('刀盾',freq=20000)
jieba.add_word('歪比',freq=20000)
jieba.add_word('巴卜',freq=20000)
```

```python
jieba.lcut('比比拉布', HMM=False)
```

```python
jieba.lcut('我的刀盾', HMM=False)
```

```python
jieba.lcut('歪比巴卜', HMM=False)
```

同样是长词词频小短词词频大，歪比巴卜最终按短词划分，我的刀盾按长词划分

原因是P=P1 * P2 * p3 * ... 最大概率算法中乘积会让最终结果概率更小，天然地长词划分在路径选择上占优势
