```python
import jieba
import jieba.posseg as pseg
```

```python
text = "九头虫让奔波儿灞把唐僧师徒除掉"
```

```python
jieba.load_userdict('./user_dict.txt')
```

```python
# HMM=False 强制只使用词典和动态规划
words = pseg.lcut(text, HMM=False)
```

```python
print(f"默认词性输出: {words}")
```

九头虫和奔波儿灞的词性没能正确标注为名词，是因为用户词典未规定词性
