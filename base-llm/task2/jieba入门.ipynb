```python
import jieba
```

基于词典构建 DAG 并利用动态规划寻找最大概率路径的方法，在 jieba 中对应的就是精确模式

```python
text = '我在梦里回到了大唐'
```

```python
seg_list = jieba.lcut(text, cut_all=False) # cut_all=False 表示精确模式
```

```python
print(seg_list)
```

DAG路径方案的前提是每个seg都在词典中，这里的奔波儿灞显然不在词典中

```python
text = "九头虫让奔波儿灞把唐僧师徒除掉"
print(f"精准模式: {jieba.lcut(text, cut_all=False)}")
```
