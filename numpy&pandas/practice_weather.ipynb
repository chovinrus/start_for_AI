分组聚合处理

```python
df['month'] = pd.to_datetime(df['date']).dt.to_period('M').astype(str)
df
```

```python
df.groupby('month') # 返回一个分组对象
```

```python
df.groupby('month')[['temp_max','temp_min']] #返回分组对象部分列
```

```python
df.groupby('month')[['temp_max','temp_min']].mean()
```

```python
df.groupby('month')['weather'].nunique()
```

```python
df.groupby('month')[['temp_min', 'temp_max']].mean().plot()
```
