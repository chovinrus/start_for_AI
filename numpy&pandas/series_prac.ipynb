```python
import pandas as pd
```

索引在左边，值在右边。由于我们没有为数据指定索引，于是会自动创建一个0到N-1（N为数据的长度）的整数型索引。

```python
s = pd.Series([4, 7,-5, 3])
print(s)
```

手动指定索引

```python
s = pd.Series([4, 7,-5, 3], index=["a", "b", "c", "d"])
print(s)
```

指定索引名称

```python
s = pd.Series([1,2,3,4], index=[2,3,4,5], name='my_index')
print(s)
```

通过字典创建Series

```python
d = {'大帅比':'朱建民', '大美女':'林黛玉'}
s = pd.Series(d)
print(s)
```

```python
s = pd.Series(d, index=['大帅比'])
print(s)
```

Series的常用属性

```python
d = {'大帅比':'朱建民', '大美女':'林黛玉'}
s = pd.Series(d, name='帅哥美女名录')
print(s.index)
print(s.values)
print(s.ndim) #维度
print(s.shape)
print(s.size)
print(s.dtype)
print(s.name)
print(s.loc['大帅比']) #显示索引，即实际用户添加的索引
print(s.iloc[1]) #显示索引，即实际用户添加的索引
print(s.loc[['大帅比','大美女']]) #loc支持按索引数组查询
print(s.at["大帅比"]) #at只支持查询单个元素
print(s.iat[0]) #iat只支持查询单个元素
```

其他常见的方法

```python
d = {'大帅比':'朱建民', '大美女':'林黛玉'}
s = pd.Series(d, name='帅哥美女名录')
print(s.head(1)) # 查看前n行，不传参数默认为5行
print(s.tail(1)) # 查看前n行，不传参数默认为5行
print(s.isin(['朱建民','林黛玉'])) 
print(s.isna())  #查看所有元素中是否为非nan或none
print(s.sum())
print(s.min())
print(s.max())
d = {'大帅比':1, '大美女':2}
s = pd.Series(d, name='帅哥美女名录')
print(s.mean())
print(s.median()) #中位数
print(s.mode) #众数
#分位数，pandas.Series.quantile 方法中的 interpolation 参数，其作用就是指定当计算的分位数位置恰好落在两个数据点之间时，如何确定最终的数值，默认是linear
print(s.quantile(0.25, interpolation='linear')) 
import numpy as np
d = {'大帅比':1, '大美女':2, 'nan值':np.nan}
s = pd.Series(d, name='帅哥美女名录')
print(s.count()) #统计非缺失值的元素数量
print(s.value_counts()) #统计所有元素出现的次数,返回的也是一个series
print(len(s))
s = pd.Series([2,3,2,4,5])
print(s.unique()) #去重，返回ndarray
print(s.drop_duplicates()) #去重返回series
print(s) #原series无变化
print(s.nunique()) #不重复元素的数量
s.index=[3,7,1,2,0]
s = s.sort_index() #按索引排序,返回的是副本，不是就地排序
print(s)
s = s.sort_values() #按值排序,返回副本，同样不是就地排序
print(s)
s = s.replace(2,6) #替换元素，返回副本
print(s)
s_ = pd.Series(s)
print(s.equals(s_)) #判断两个series内容是否相同
print(id(s_)) 
print(id(s))
print(s.keys()) #返回s的index
print(s.index.equals(s.keys()))
for i in s.items(): # 返回索引和值的zip元组
    print(i)
print(s.corr(s_)) #皮尔逊相关性系数
print(s.cov(s_)) #计算协方差
s.hist() #根据元素出现频次绘制直方图
```

series的布尔运算和布尔索引

```python
bool_ = s > s.mean() #series 和标量做计算得到元素为布尔类型的series
print(bool_)
print(s[bool_]) #用[]接收索引相同的布尔类型series，返回一个筛选后的series
```

series运算

```python
print(s * 2) #与标量运算
print(s + pd.Series([1,2,3])) #与series运算，数量不匹配自动在少的一方补充nan
```
