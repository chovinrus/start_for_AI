 zeros()、ones()、empty()与zeros_like()、ones_like()、empty_like()

 zeros()：返回给定形状和类型的新数组，用0填充。
ones()：返回给定形状和类型的新数组，用1填充。
empty()：返回给定形状和类型的未初始化的新数组。
需要注意的是，np.empty并不保证数组元素被初始化为0，它只是分配内存空间，数组
中的元素值是未初始化的，可能是内存中的任意值。
上述3个方法创建的数组元素类型默认都是float64。

zeros_like()：返回与给定数组具有相同形状和类型的0新数组。
ones_like()：返回与给定数组具有相同形状和类型的1新数组。
empty_like()：返回与给定数组具有相同形状和类型的未初始化的新数组

```python
arr1 = np.zeros((2, 5))
print(arr1)
arr2 = np.empty((2,4))
print(arr2)
arr3 = np.ones_like(arr1)
print(arr3)
```

full()：返回给定形状和类型的新数组，用指定的值填充。

full_like()：返回与给定数组具有相同形状和类型的用指定值填充的新数组

```python
arr1 = np.full((2, 3), 6)
arr2 = np.full_like(arr1,5)
print(arr1)
print(arr2)
```

等差数列的endpoint

以上等差数列的案例可能看着有点懵，核心是np等差数列返回元素的依据是num参数，返回数组包括了end值，因此等差间隔数量是num-1。如果希望等差间隔数量和num一致，返回数组应该是左开右闭即不包含end，此时需要设置endpoint为False（默认是True）

```python
np.linspace(0,10,2, endpoint=False)
```

#### ndarray上的操作

ndarray对索引与切片的支持和list一样，在这里不再赘述

ndarray支持的函数有abs、ceil、floor、rint（四舍五入）、isnan、multiply、divide、where（条件运算符）

```python
arr1 = np.random.randn(2,3)
print(arr1)
print(np.abs(arr1))
print(np.ceil(arr1))
print(np.floor(arr1))
print(np.rint(arr1))
print(np.isnan(arr1))
print(np.multiply(arr1, 2))
print(np.divide(arr1, arr1))
print(np.where(arr1 > 0, 1, 0))
```

聚合函数、统计函数

```python
arr1 = np.random.randint(1, 5, (2, 3))
print(arr1)
print(np.mean(arr1))
print(np.sum(arr1))
print(np.max(arr1))
print(np.min(arr1))
print(np.std(arr1)) #标准差
print(np.var(arr1)) #方差
print(np.argmax(arr1)) #最大值的下标索引值
print(np.argmin(arr1))
print(np.cumsum(arr1)) #返回一个一维数组，每个元素都是之前所有元素的累加和
print(np.cumprod(arr1)) #返回一个一维数组，每个元素都是之前所有元素的累计乘积
print(np.cumprod(arr1, axis=1)) #axis：指定沿着哪个轴计算累积乘积。默认为 None，此时会将数组展平后再计算
```

numpy 中的数组不用编写循环即可执行批量运算，称之为矢量化运算。
大小相等的数组之间的任何算术运算都会将运算应用到元素级

```python
arr1 = np.array([[1, 2, 3], [4, 5, 6]])
arr2 = np.array([[7, 8, 9], [10, 11, 12]])
print(arr1 + arr2)
print(arr1- arr2)
print(arr1 * arr2)
print(arr1 / arr2)
```

数组与标量的算术运算会将标量值传播到各个元素

```python
arr1 = np.array([[1, 2, 3], [4, 5, 6]])
print(arr1 + 100)
print(arr1- 100)
print(arr1 * 100)
print(arr1 / 100)
```

ndarray之间支持运算，前提是shape兼容
- 维度允许不一致，小维度的数组通过在低纬度添加维度的形式对齐维度，如(3)要和(3,4,5)对齐维度，原数组的shape要变为(1,1,3)
- 接下来从高维度到低纬度检查兼容性，如果当前维度大小相同或其中一个大小为1，则互相兼容可以运算，否则不兼容，不能计算
- 维度大小其中一方为1的情形需“扩展”后才能真正兼容，比如[[1]]和[[],[],[]]，前者的大小为1.扩展后为[[1],[1],[1]]。复制的是当前维度的完整元素内容
- 此时两者的shape完全一致，按位置逐个进行运算即可
- 值得注意的是，ndarray的  

```python
arr1 = np.array([1, 2, 3]) 
arr2 = np.array([[4], [5], [6]])
# arr1的shape为(3,)，arr2的shape为(3,1)，为对齐维度，arr1变为[[1,2,3]]
# 从高维到低维度继续处理shape
# 维度2的大小分别为3,1，arr2扩展后变为[[4,4,4],[5,5,5],[6,6,6]]
# 维度1的大小分别为1,3，arr1扩展后变为[[1,2,3],[1,2,3],[1,2,3]]
result = arr1 + arr2
print("规则1示例结果：\n", result)
```

矩阵乘法运算不是用*或者multiply而是dot方法或内置函数dot、@运算符

```python
import numpy as np

# 设置小参数
batch_size = 2
seq_len = 3
d_model = 4
d_k = 2

# 随机生成输入 X (batch_size, seq_len, d_model)
X = np.random.randint(1, 5, size=(batch_size, seq_len, d_model)).astype(float)
# 权重 W (d_model, d_k)
W = np.random.randint(1, 3, size=(d_model, d_k)).astype(float)

print("X 形状:", X.shape)
print("X:\n", X)
print("W 形状:", W.shape)
print("W:\n", W)

# 批量矩阵乘法
Y = X @ W  # 或者 np.matmul(X, W)
print("Y 形状:", Y.shape)
print("Y:\n", Y)

# 手动验证：用循环计算第一个样本的第一个位置
i, j = 0, 0  # 第一个样本的第一个位置
manual_y = np.dot(X[i, j], W)  # 向量 (d_model,) 乘矩阵 (d_model, d_k) -> (d_k,)
print("手动计算 X[0,0] @ W =", manual_y)
print("Y[0,0] =", Y[0,0])
```
